NaN (Not a Number):

Explanation: This is a standard placeholder in numerical datasets, particularly in programming languages like Python (NumPy/Pandas). It signifies that a particular value is undefined or unrepresentable, typically used for missing numerical data.
N/A (Not Applicable):

Explanation: This term indicates that a value is not applicable in the context of the data. It is often used in categorical datasets where certain entries may not have relevant data points.
Null:

Explanation: In databases and data analysis, "null" represents the absence of a value or an unknown value. It's similar to NaN but can also apply to non-numerical data types.
None:

Explanation: In Python, None is a special constant that represents the absence of a value or a null value. It's commonly used in data structures to indicate missing or non-existent data.
Missing:

Explanation: This term generally refers to any data point that is absent in the dataset. It can be represented by various placeholders (like NaN, N/A, etc.) and needs to be addressed during data cleaning.
Empty:

Explanation: An empty value often refers to a string or a data entry that has no content (e.g., ""). It's important to identify and handle empty values as they can affect data analysis and processing.
Outlier:

Explanation: An outlier is a data point that differs significantly from other observations in a dataset. It may indicate variability in the measurements or might suggest a data entry error. Identifying and treating outliers is crucial during data cleaning.
Duplicate:

Explanation: Duplicate records occur when the same data entry appears more than once in a dataset. These can skew analysis results and should be identified and removed.
Standardization:

Explanation: This refers to the process of bringing different data formats or values into a common format. For example, dates might be standardized to a single format (e.g., YYYY-MM-DD).
Normalization:

Explanation: Normalization involves adjusting values in a dataset to a common scale, often used for numerical data to improve model performance in machine learning. This can include min-max scaling or z-score normalization.

#  Imputing Null values
mean = avg, median = enter term , mode = frequently used 

Impute Median value for Age column

Impute Mode for Type of Contract

Impute Median for Duration of Pitch

Impute Mode for NumberofFollowup as it is Discrete feature

Impute Mode for PreferredPropertyStar

Impute Median for NumberofTrips

Impute Mode for NumberOfChildrenVisiting

Impute Median for MonthlyIncome

```python
#Age
df.Age.fillna(df.Age.median(), inplace=True)

#TypeofContract
df.TypeofContact.fillna(df.TypeofContact.mode()[0], inplace=True)

#DurationOfPitch
df.DurationOfPitch.fillna(df.DurationOfPitch.median(), inplace=True)

#NumberOfFollowups
df.NumberOfFollowups.fillna(df.NumberOfFollowups.mode()[0], inplace=True)

#PreferredPropertyStar
df.PreferredPropertyStar.fillna(df.PreferredPropertyStar.mode()[0], inplace=True)

#NumberOfTrips
df.NumberOfTrips.fillna(df.NumberOfTrips.median(), inplace=True)

#NumberOfChildrenVisiting
df.NumberOfChildrenVisiting.fillna(df.NumberOfChildrenVisiting.mode()[0], inplace=True)

#MonthlyIncome
df.MonthlyIncome.fillna(df.MonthlyIncome.median(), inplace=True)
```
